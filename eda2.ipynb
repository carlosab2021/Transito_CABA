Realizamos un análisis exploratorio de datos (EDA) de dos tablas, "hechos" y "victimas", contenidas en el archivo llamado "homicidios.xlsx".
 
Importamos las bibliotecas necesarias: Importamos las bibliotecas pandas, matplotlib.pyplot y seaborn para el análisis de datos y la visualización.

Cargamos el archivo "homicidios.xlsx": Utilizamos pandas para cargar las dos hojas de datos del archivo Excel en dos DataFrames llamados "df_hechos" y "df_victimas".

Información básica: Mostramos información básica sobre los DataFrames, como la cantidad de filas y columnas, así como los tipos de datos en cada columna. Esto nos ayuda a entender la estructura de los datos.

Estadísticas descriptivas: Calculamos estadísticas descriptivas, como la media, la desviación estándar y los percentiles, para las variables numéricas en ambos DataFrames. Esto nos proporciona una idea de la distribución y las tendencias en los datos.

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo "homicidios.xlsx"
df_hechos = pd.read_excel('homicidios.xlsx', sheet_name='HECHOS')
df_victimas = pd.read_excel('homicidios.xlsx', sheet_name='VICTIMAS')

# Comenzar con el análisis exploratorio de datos (EDA)
# Análisis de las tablas "hechos" y "victimas"
# ...

# Verificar la información básica de los DataFrames
print("Información básica de la tabla hechos:")
print(df_hechos.info())
print("\nInformación básica de la tabla victimas:")
print(df_victimas.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas de la tabla hechos:")
print(df_hechos.describe())
print("\nEstadísticas descriptivas de la tabla victimas:")
print(df_victimas.describe())

# Visualizar la distribución de variables numéricas
sns.histplot(df_hechos['N_VICTIMAS'], bins=10, kde=True)
plt.title("Distribución de N_VICTIMAS en la tabla hechos")
plt.xlabel("Número de Víctimas")
plt.ylabel("Frecuencia")
plt.show()

# Búsqueda de valores faltantes
print("\nValores faltantes en la tabla hechos:")
print(df_hechos.isnull().sum())
print("\nValores faltantes en la tabla victimas:")
print(df_victimas.isnull().sum())


# Eliminar registros duplicados (si los hay)
df_hechos = df_hechos.drop_duplicates()
df_victimas = df_victimas.drop_duplicates()

# Conclusión y análisis final
# ...

# Guardar los DataFrames limpios y listos para su uso posterior
df_hechos.to_csv('hechos_limpios.csv', index=False)
df_victimas.to_csv('victimas_limpios.csv', index=False)
```


Los valores atípicos identificados en el conjunto de datos corresponden a un total de 19 accidentes en los cuales se registraron 2 víctimas fatales. Además, se destaca un accidente excepcional donde se reportaron 3 víctimas fatales. Este accidente particular está registrado bajo el ID 2017-0035 y ocurrió el 23 de marzo de 2017 a las 05:00:00 horas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Cargar los datos
df_hechos = pd.read_excel('homicidios.xlsx', sheet_name='HECHOS')

# Calcular el Z-score
z_scores = stats.zscore(df_hechos['N_VICTIMAS'])
threshold = 3  # Umbral para considerar valores atípicos

# Identificar los valores atípicos
outliers = df_hechos[abs(z_scores) > threshold]

# Crear un gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_hechos.index, y=df_hechos['N_VICTIMAS'], label="Datos")
sns.scatterplot(x=outliers.index, y=outliers['N_VICTIMAS'], color='red', marker='X', s=100, label="Valores Atípicos")

plt.title("Gráfico de Dispersión de N_VICTIMAS con Valores Atípicos")
plt.xlabel("Índice")
plt.ylabel("N_VICTIMAS")
plt.legend()

# Configurar el eje Y para mostrar solo valores enteros
plt.yticks(range(int(df_hechos['N_VICTIMAS'].max()) + 1))

plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# Cargar la base de datos
df_hechos = pd.read_excel('homicidios.xlsx', sheet_name='HECHOS')
# Agrupar categorías de PARTICIPANTES
def agrupar_participantes(categoria):
    if 'MOTO' in categoria:
        return 'Motocicletas'
    elif 'AUTO' in categoria:
        return 'Automóviles'
    elif 'PEATON' in categoria:
        return 'Peatones'
    elif 'BICICLETA' in categoria:
        return 'Bicicleta'
    elif 'CARGAS' in categoria:
        return 'Cargas'
    elif 'PASAJEROS' in categoria:
        return 'Pasajeros'
    else:
        return 'Otro'

df_hechos['PARTICIPANTES_AGRUPADOS'] = df_hechos['PARTICIPANTES'].apply(agrupar_participantes)

# Calcular estadísticas descriptivas para N_VICTIMAS por grupo de PARTICIPANTES
estadisticas_por_grupo = df_hechos.groupby('PARTICIPANTES_AGRUPADOS')['N_VICTIMAS'].describe()

# Imprimir las estadísticas descriptivas
print(estadisticas_por_grupo)

```

Unimos las tablas Hechos y Victimas.

```python
import pandas as pd

# Cargar los DataFrames de Hechos y Víctimas
df_hechos = pd.read_excel('homicidios.xlsx', sheet_name='HECHOS')
df_victimas = pd.read_excel('homicidios.xlsx', sheet_name='VICTIMAS')

# Fusionar los DataFrames utilizando la columna ID en Hechos y ID_hechos en Víctimas
df_completo = pd.merge(df_hechos, df_victimas, left_on='ID', right_on='ID_hecho', how='inner')

# Mostrar el DataFrame resultante que contiene la relación entre las dos tablas
print(df_completo.head())

# Guardar el DataFrame resultante si es necesario
df_completo.to_csv('homicidios_completo.csv', index=False)
df_completo.to_excel('homicidios_completo.xlsx', index=False)
```
